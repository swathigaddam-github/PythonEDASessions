```python
** task-1  **
~  I want to create two lists
~  1. categorical list
~  2. Numerical column list
~   using dtypes
     ~  whenever you want to create any list from series
     ~  first conver a series into a dictionary
     ~ when you convert it become a key : value pair
~ iterate through dictionary
~ if that value== object type, then keep that corresponding key in one list
~ else keep that variables in other list
```

```python
data_types=visa_df.dtypes
dict(data_types)
```

```python
#dict.keys(data_types)
categorical_list=[]
num_list=[]
for i,j in dict(data_types).items():
    #print(i,j)
    if j=='object':
        categorical_list.append(i)
    else:
        num_list.append(i)
    
```

```python
for i,j in dict(data_types).items():
    print(i,j)
```

```python
categorical_list=[i for i,j in dict(data_types).items() if j=='object']
num_list=[i for i,j in dict(data_types).items() if j!='object']
categorical_list,num_list
```
