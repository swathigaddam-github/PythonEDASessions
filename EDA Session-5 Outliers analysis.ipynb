```python
# step-1: Make first condition continent=='Asia'
# Step-2 : Make the Second condition case_status=='certified'
# step-3 : Make the third condition : con1& con2
# step-4 : Then retrieve dataframe and take the count
```

```python
continents=visa_df['continent'].unique()
certified_list=[]
denied_list=[]
for i in continents:
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=="Certified"
    con3=visa_df['case_status']=="Denied"
    cert_con=con1&con2
    den_con=con1&con3
    cert_count=len(visa_df[cert_con])
    den_count=len(visa_df[den_con])
    #count=len(visa_df[con3])
    print(f"no.of certified visa from {i} is : {cert_count}")
    print(f"no.of denied visa from {i} is : {den_count}")
    certified_list.append(cert_count)
    denied_list.append(den_count)
    
```

```python
~ implement covariance function
~ Covariance vs correlation
    ~ Covariance will tell about columns have rlation or not
    ~ Covariance will explain about relation status
        ~ Positive relation
        ~ negative relation
        ~ no relation
    ~ Correlation tells about how much relation between two variables
    ~ It denotes with 'r'
    ~ r varies from -1 to 1
    ~ -1 to 0 indicates negative relation quantity
    ~ 0 to 1 indicates positive relation quantity
    ~ 0 indicates no relation
```

$method$ $name$:   $corr$

```python
visa_df.corr(numeric_only=True)#  1,1,1 in diagoanl i.e   variance
```

 $ HEAT $  $MAP$

```python
~ Heat map is one of the important visualization method, to show the matrix 
~ It is under Seaborn package
~ In every matrix we have heighest values and lowest values available
~ Heat map will provide colour visualiztion
~ Also it provides color bar, which colour indicates which value
```

```python
corr_data=visa_df.corr(numeric_only=True)#  1,1,1 in diagoanl i.e   variance
sns.heatmap(corr_data,annot=True)# annotations will give color numbers
plt.show()
```

```python
corr_data=visa_df.corr(numeric_only=True)#  1,1,1 in diagoanl i.e   variance
sns.heatmap(corr_data,annot=True,cmap='ma')# annotations will give color numbers
plt.show()
```

```python
~ By seeing the Graphs we might not say the relation status
~  It is best practice get the percentage of relation using correlation
~ All 3 columns having approximately 0 percen of relation
~ which indicates all columns having no relation
~ no relation means independent each other
~ In ML we expect input columns should be independent each other
~ Correlation method gives matrix view
~ We have 3 columns 3*3 total 9 fields
~ If a data has 20 columns then 20*20 total 400 fields
~ thats why always go with Heat map visualization
```
