# $Map$


```python
~ Map is one of method to convert categorical values to numerical column 
~ Take one Categorical column ~ Get unique labels first 
~ make a dictionary with assign a number to each label 
~ foe examplecase_status has two labels
    ~ Certified 
    ~ Denied
~Assign 0 to certified 1 to denied
~ Createe a dictionary labels as keys , number as values only
~ {certified :0,Denied:1}
```

```python
# Read column
# Get unique labels
# Make dict
# apply map
```

```python
visa_df=pd.read_csv(path)
cat_columns=visa_df.select_dtypes(include='object').columns
for i in cat_columns:
    unique_labels=sorted(visa_df[i].unique())
    list1=[i for i in range(len(unique_labels))]
    dict1=dict(zip(unique_labels,list1))
    dict1
    visa_df[i]=visa_df[i].map(dict1)
visa_df
```

```python
visa_df['case_status']=visa_df['case_status'].map(dict1)
visa_df
```

$ Label$ $Encoder$

```python
~ Label Encoder is a method to convert categorical data to Numerical data
~ LabelEncoder is a same way,how we did in above
~ The for loop map method we explored
~ But Label Encoder also do same
~ It is under  Sickit learn we usually called as sklearn
~ Under skleran we have class called as preprocessing
~ Because currently we are doing data preprocessing only
~ under preprocesiing class we have a method callesas LableEncoder
~ sklearn 
    ~ Preprocessing
    ~ LabelEncoder
~ any sklearn package has 3 steps
        step-1:  Read package
        step-2: Save package
        step-3: Apply fit transform on data
```

 $fit$ $and$ $ transform$

```python
~ fit means develop the logic
~ transform means process the logic
~ If we observe in map method first we developed dictionary logic
~ This  we apply the  dict to column
~ Developing dict is kind of fit
~ Apply dict to column means we are transforming data cat to num
~ whenever we are transforming or changing data then use fit transform
~ If we are not changing data then only fit
```

$np.where$

```python
~ using np.where also we can change the Categorical data to Numerical 
~ But np.where is only for binary condition
~ np.where is same as if-else
~ if condition is True:will apply True value
~ if condition is False:will apply the False value
~ So np.where use for only binary labels
~ A column should have only two unique labels
```

```python
visa_df=pd.read_csv(path)
```

```python
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```

$ One$ $hot$ $Encoder$

~ ON represents with 1
~ OFF represents with 0

~  One hot Encoder creates new columns which is equal to number of unique labels 
~ example: case status will create two extra columns
    ~ case_status_Certified
    ~ case_status_Denied
|case_status|case_status_certified|case_status_denied|
